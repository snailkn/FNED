```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('data/train_7000.json', 'r') as f:
    tr_data = json.load(f)
with open('data/valid_1500.json', 'r') as f:
    val_data = json.load(f)
with open('data/test_dataset_A.json', 'r') as f:
    test_data = json.load(f)
```

```python
triggers, arguments, entities, no_event_items = [], [], [], []
for item in tr_data + val_data:
    event = item['event_mention']
    entities.extend(item['entity_mention'])
    if event:
        event_type = event['event_type']
        trigger = event['trigger']['text']
        start, end = event['trigger']['offset']
        triggers.append((trigger, event_type, end - start))
        for arg in event['arguments']:
            arguments.append((event_type, trigger, arg['role'], arg['text']))
    else:
        no_event_items.append(item)
```

```python
df_trigger = pd.DataFrame(triggers, columns=['content', 'event_type', 'trigger_len'])
df_arg = pd.DataFrame(arguments, columns=['event_type', 'trigger', 'role', 'content'])
```
